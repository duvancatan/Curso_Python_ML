```python
cols=["customerID","gender", "tenure", "MonthlyCharges", "TotalCharges"]
```

```python
gf = df[cols]
gf.head()
```

```python
#CONSTRUCCIÓN DE CAMPOS
gf = gf.assign(Percentage = lambda x : (2*x["MonthlyCharges"]))
gf.head()
```

```python
#CONSTRUCCIÓN DE CAMPOS A PARTIR DE VARIOS CAMPOS
gf = gf.assign(Suma = lambda x : (x["MonthlyCharges"] / x["tenure"]))
gf.head()
```

```python
#ETIQUETAR CON 0 Y 1 EL GÉNERO (IMPORTANTE PARA LA CLASIFICACIÓN)
cod_to_int = {
    'Male': 1,
    'Female': 0
}
sub_df = gf.assign(label = lambda x: x.gender.map(cod_to_int))
```

```python
sub_df
```

```python
#AGRUPACIÓN 
sub_df.groupby("gender").sum()
```

```python
#AGRUPACIÓN CON DISTINTOS VALORES POR CAMPO (AGREGACIONES)
grouped = sub_df.groupby("gender").agg({
    "tenure": "max",
    "MonthlyCharges": "mean",
    "Percentage": "min"
})
grouped
```

```python
import matplotlib.pyplot as plt
```

```python
grouped["MonthlyCharges"].plot()

plt.show()
```

```python
grouped["MonthlyCharges"].plot(kind="bar")

plt.show()
```
