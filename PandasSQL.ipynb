```python
from pandasql import sqldf
import pandas as pd
from sklearn.datasets import load_iris
from pandasql import sqldf
from pandasql import load_meat, load_births
import re
```

```python
#Cargar las tres datas
births = load_births()
meat = load_meat()
iris = load_iris()
```

```python
#creamos dataframe (es una estructura de datos tabular bidimensional de tamaño variable y potencialmente heterogénea con ejes etiquetados (filas y columnas).)
iris_df = pd.DataFrame(iris.data, columns=["Petal length","Petal Width","Sepal Length","Sepal Width"])
```

## Categorical. from_codes.
Haga un tipo categórico a partir de códigos y categorías o dtype. Este constructor es útil si ya tiene códigos y categorías / dtype y,por lo tanto, no necesita el paso de factorización (computación intensiva), que generalmente se realiza en el constructor.


```python
iris_df
```

```python
#‘target_names’, the meaning of the labels
#Si necesita la etiqueta de cadena puede usar target_names y agregue una nueva columna llamada species:
iris_df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
```

```python
iris_df
```

```python
#Salvavidas
iris_df.columns = [re.sub("[() ]", "", col) for col in iris_df.columns]
```

```python
print(sqldf("SELECT * FROM iris_df LIMIT 10;"))
```

```python
#Agrupar por agregacion promedio, minimo y máximo
#min
q = """
      select
        species
        , avg(SepalWidth)
        , min(SepalWidth)
        , max(SepalWidth)
      from
        iris_df
      group by
        species;
        
"""
```

## Project description
pysqldf le permite consultar pandas DataFrames utilizando la sintaxis SQL. Funciona de manera similar a sqldf en R. pysqldf busca proporcionar una forma más familiar de manipular y limpiar datos para personas nuevas en Python o pandas.

```python
print("*" * 80)
print("aggregation")
print("-" * 80)
# Imprime q
print(q)
#promedio, min y máximp tres tipo de especies sectosa, versicolor, virginica
print(sqldf(q, locals()))
def pysqldf(q):
    "agregue esto a su script si se cansa de llamar a los locales ()"
    return sqldf(q, globals())
print("-" * 80)
print(q)
print(pysqldf(q))
```

## #Inner Join

```python
q = """
    select
        a.*
    from
        iris_df a
    inner join
        iris_df b
            on a.species = b.species
    limit 10;
"""
print("*" * 80)
print("joins")
print("-" * 80)
print(q)
print(pysqldf(q))
```

## where

```python
q = """
    select
        *
    from
        iris_df
    where
        species = 'virginica'
        and SepalLength > 7.7;
"""
print("*" * 80)
print("where clause")
print("-" * 80)
print(q)
print(pysqldf(q))
iris_df['id'] = range(len(iris_df))
q = """
    select
        *
    from
        iris_df
    where
        id in (select id from iris_df where SepalLength*SepalLength > 25);
"""
print("*" * 80)
print("subqueries")
print("-" * 80)
print(q)
print(pysqldf(q))
```

## INNER JOIN

```python
q = """
    SELECT
        m.*
        , b.births
    FROM
        meat m
    INNER JOIN
        births b
            on m.date = b.date
    ORDER BY
        m.date;
"""

print(pysqldf(q).head())
```
